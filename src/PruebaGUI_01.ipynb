```julia
# Para convertir a Cadena de Caracteres

function insert_data(input_str::AbstractString)
cadena = Vector{Char}()
    for char in input_str
        push!(cadena,char)
    end
        return cadena
end

input_str = "Hola Mundo"
cadena = insert_data(input_str)
println(cadena)
```
